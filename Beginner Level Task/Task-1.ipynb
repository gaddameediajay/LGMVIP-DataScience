```python
import cv2
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
image = cv2.imread("pic2.jpg")
cv2.imshow("Nature", image)
cv2.waitKey(0)
```

```python
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
plt.imshow(gray_image,cmap="gray")
plt.title("Black And White Nature")
```

```python
inverted_image = 255 - gray_image
plt.imshow(inverted_image,cmap="gray")
plt.title("Black And White Inverted")
```

```python
blurred = cv2.GaussianBlur(inverted_image, (21, 21), 0)
inverted_blurred = 255 - blurred
plt.imshow(inverted_blurred,cmap="gray")
plt.title("Inverted Blurred GrayScale Image")
```

```python
pencil_sketch = cv2.divide(gray_image, inverted_blurred, scale=256.0)
plt.imshow(pencil_sketch,cmap="gray")
plt.title("Inverted Blurred GrayScale Image")
```

```python
cv2.imshow("Nature Original", image)
cv2.imshow("Nature Pencil Sketch", pencil_sketch)
cv2.waitKey(0)
```
